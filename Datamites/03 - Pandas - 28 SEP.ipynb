```python
for i,dur in enumerate(data.Duration):
    dur  = dur.split()
    if len(dur) == 2:
        data.loc[i,'dur_mins'] = int(dur[0].replace('h',''))*60+int(dur[1].replace('m',''))
    elif 'h' in dur[0]:
        data.loc[i,'dur_mins'] = int(dur[0].replace('h',''))*60
    elif 'm' in dur[0]:
        data.loc[i,'dur_mins'] = int(dur[0].replace('m',''))
```
